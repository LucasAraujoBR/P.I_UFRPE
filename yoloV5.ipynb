# Teste com uma imagem

```python
import torch
import cv2

# Carregar o modelo treinado
model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/train/exp16/weights/best.pt')

# Carregar uma imagem
img = cv2.imread('dataset/images/frame_000001.PNG')

# Fazer inferência
results = model(img)

# Exibir resultados
results.print()  # exibir os resultados no console
results.show()  # mostrar a imagem com as caixas delimitadoras

# Extrair caixas delimitadoras e contagem de carros
car_count = len(results.xyxy[0])  # número de carros detectados
print(f"Total de carros: {car_count}")
```

# Teste com várias imagens

```python
import os
import torch
import cv2

# Carregar o modelo treinado
model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/train/exp6/weights/best.pt')

# Caminho para o diretório de imagens
image_dir = 'dataset/images/'

# Iterar pelas imagens
for image_file in os.listdir(image_dir):
    img_path = os.path.join(image_dir, image_file)
    img = cv2.imread(img_path)

    # Fazer inferência
    results = model(img)
    
    # Contar carros
    car_count = len(results.xyxy[0])
    
    print(f"Imagem: {image_file}, Total de carros: {car_count}")
```
